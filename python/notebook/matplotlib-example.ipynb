```python
import numpy as np

list = [[1, 3, 5], [2, 4, 6]]
np_array = np.array(list, dtype=float)
np_matrix = np.mat(list, dtype=float)
print('np_array type is {}, np_matrix type is {}'.format(type(np_array), type(np_matrix)))

merged_arr = np_array + np_matrix
print('merged_arr type is {}, value is {}'.format(type(merged_arr), merged_arr))
```
